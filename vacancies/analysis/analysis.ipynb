```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

technologies = np.array([
    "Python", "Django", "FastAPI", "Flask", "Docker", "Kubernetes",
    "PostgreSQL", "MySQL", "MongoDB", "Redis", "Numpy", "Pandas",
    "Tableau", "PowerBI", "TensorFlow", "PyTorch", "AWS", "Azure",
    "GCP", "Celery", "RabbitMQ", "Git"
])
```

A Function to Count Technology Mentions in Descriptions

```python
def count_technologies_np(descriptions, technologies):
    technologies = np.char.lower(technologies)
    achievements = np.zeros(len(technologies), dtype=int)
    
    for description in descriptions:
        description_lower = description.lower()
        for i, tech in enumerate(technologies):
            if tech in description_lower:
                achievements[i] += 1
                continue
    
    return achievements
```

A Function to Draw bar

```python
def draw_bar(tech_counts_df, title: str):
    plt.figure(figsize=(12, 6))
    plt.bar(tech_counts_df["Technology"], tech_counts_df["Percentage"], color="skyblue")
    plt.title(title, fontsize=16)
    plt.xlabel("Technology", fontsize=14)
    plt.ylabel("Percentage (%)", fontsize=14)
    plt.xticks(rotation=45, ha="right")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()
```

Analysis of Technologies for Junior-Middle Level Vacancies


```python
jun_mid_df = pd.read_csv("../../python_jobs_exp_1_3.csv")

tech_counts = count_technologies_np(jun_mid_df["description"], technologies)

tech_counts_df = pd.DataFrame({
    "Technology": technologies,
    "Count": tech_counts,
    "Percentage": (tech_counts / len(jun_mid_df["description"])) * 100
})

tech_counts_df = tech_counts_df.sort_values(by="Count", ascending=False)

draw_bar(tech_counts_df, "Most Demanded Technologies for Junior-Middle Python Developers")
```

Analysis of Technologies for Middle-Senior Level Vacancies


```python
jun_mid_df = pd.read_csv("../../python_jobs_exp_3_5.csv")

tech_counts = count_technologies_np(jun_mid_df["description"], technologies)

tech_counts_df = pd.DataFrame({
    "Technology": technologies,
    "Count": tech_counts,
    "Percentage": (tech_counts / len(jun_mid_df["description"])) * 100
})

tech_counts_df = tech_counts_df.sort_values(by="Count", ascending=False)

draw_bar(tech_counts_df, "Most Demanded Technologies for Middle-Senior Python Developers")
```
